```python
from bs4 import BeautifulSoup

# HTTP GET Requests
req = requests.get('https://beomi.github.io/beomi.github.io_old/')

# HTML 소스 가져오기
html = req.text

# BeautifulSoup으로 html 소스를 python 객체로 변환하기
# 첫 인자: html 소스코드, 두 번째 인자: 어떤 parser를 이용할지 명시
# 여기에서는 Python 내장 html.parser를 이용함
soup = BeautifulSoup(html, 'html.parser')
```

## 웹페이지 살펴보기
* 블로그의 모든 제목을 가져오기
* 크롬에 내장된 검사도구를 이용 -> 현재 title은 a 태그로 구성되어 있음을 알 수 있음
* 제목은 h3 tag 밑에 a tag 안에 쌓여있는 것을 알 수 있음


```python
# CSS Seleector를 통해 html 요소를 찾아내기
my_titles = soup.select(
    'h3 > a'
)
my_titles
```

```python
# my_titles는 list 객체
for title in my_titles:
    # Tag 안의 텍스트
    print(title.text)
```

```python
# my_titles는 list 객체
for title in my_titles:
    # Tag의 속성을 가져오기 (ex: href 속성)
    print(title.get('href'))
```

# 보안뉴스 기사제목 추출 예제

```python
import requests
import bs4
```

```python
# 해당 주소로 요청을 보낸 후 정보를 req 변수에 저장
req = requests.get('https://www.boannews.com/media/t_list.asp')

# req에 들어있는 html 코드를 보기 위해서 req.text로 볼 수 있음
print(req.text)
```

```python
# 원하는 부분을 쉽게 찾기 위해 bs4를 이용하기
parse = bs4.BeautifulSoup(req.text, 'html.parser')
```

```python
# <div id='news_area'> 부분에 기사가 있으므로 다음과 같이 코드 작성
newspaper = parse.find('div', {'id': 'news_area'})
print(newspaper)
```

```python
# <div id='news_list'> ~~ </div> 하나의 기사가 들어가있는 부분
newspaper = newspaper.find_all('div', {'class':'news_list'})
print(newspaper)
```

```python
# <span class="news_txt"> ~~ </span> 부분에 찾던 기사 제목이 있음
# newspaper 변수는 문자열 타입이 아니라 list 타입이므로 for 문으로 하나씩 찾아봐야 함
for string in newspaper:
    newsTitle = string.find('span', {'class':'news_txt'}).text
    print(newsTitle)
```
